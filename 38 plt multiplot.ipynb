```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def example_plot(ax, fontsize=12):
    ax.plot([1,2])#value
    ax.set_xlabel('x-label',fontsize = fontsize)
    ax.set_ylabel('y-label',fontsize = fontsize)
    ax.set_title('Title',fontsize = fontsize)
    
fig,axes = plt.subplots(2,2)
fig.set_facecolor('y')
example_plot(axes[0,0])
example_plot(axes[0,1])
example_plot(axes[1,0])
example_plot(axes[1,1])

fig.tight_layout(w_pad = 10,h_pad = 5)#调整位置布局
#fig.subplot_adjust()和上述效果相同
```

```python
#自定义布局方式
ax1 = plt.subplot(221)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)
```

```python
#复杂的布局需要GridSpec
fig = plt.figure()
gs = fig.add_gridspec(3,3)
ax1 = fig.add_subplot(gs[0,0:3]) #gs[x1:x2,y1:y2] 第x1到x2行 从y1到y2列
ax2 = fig.add_subplot(gs[1,0:2])
ax3 = fig.add_subplot(gs[1:3,2])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[2,1])

fig.tight_layout()
```

```python
#调整图像大小的比例
fig = plt.figure()
width = (1,3,1)
height = (3,3,1)

gs = fig.add_gridspec(3,3,width_ratios = width, height_ratios = height)
for row in range(0,3):
    for col in range(0,3):
        fig.add_subplot(gs[row,col])
fig.tight_layout()
```
