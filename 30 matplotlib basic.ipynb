```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager
```

```python
y = [np.random.randint(0,10) for x in range(20)] # for 是循环几次产生
x = range(5,25)

plt.plot(x,y)
```

```python
plt.plot("x","y","",data = mydata)
```

```python
mydf = pd.DataFrame(data = mydata)
plt.plot("x","y","--y",data = mydf)
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
line = lines[0]
line.set_color('r')
line.set_linewidth(4)
#设置字体
font = font_manager.FontProperties(fname=r"C:\Windows\Fonts\MI_LanTing_Regular.ttf",size =20)
plt.title("折线图",fontproperties=font)
```

```python
y3 = [np.random.randint(0,10) for x in range(10)]
plt.figure(figsize=(15,5))
plt.plot(y3, marker="o")
plt.xticks(range(10),["第%d天"%x for x in range(1,11)],fontproperties=font)#设置刻度
plt.xlabel("天数",fontproperties = font)
plt.ylabel("大小",fontproperties = font)
plt.title("图",fontproperties=font)
plt.grid()
plt.show()
```

```python
x = np.linspace(0,20,50)
y = np.sin(x)
plt.plot(x,y,marker='o',markerfacecolor='r',markersize=10,markeredgecolor='g')
```

```python
y = np.sin(np.arange(20))
plt.figure(figsize=(15,5))
plt.plot(np.sin(np.arange(20)),marker='o')
for index,value in enumerate(y):
    plt.annotate("(%.2f,%.2f)"%(index,value),xy=(index,value),xytext=(index-0.1,value-0.5),
             arrowprops={"width":10,"headwidth":16,"headlength":10,"shrink":0.3,"facecolor":'r'})
```

```python
#设置画板 figure 需要在所有画图设置前处理
plt.figure(figsize=(10,5),facecolor = 'g',edgecolor='k',linewidth = 2, dpi =200,frameon =True)
plt.plot(np.arange(5,20))
plt.savefig("x.png")
```

```python
#绘制多个图形
x = np.linspace(0,20)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
#绘制多个子图
plt.style.use("seaborn")
plt.subplot(121)
plt.plot(np.arange(20))
plt.plot(np.arange(20)**2)
plt.subplot(122)
plt.plot(np.sin(np.arange(20)),c='r')
```
