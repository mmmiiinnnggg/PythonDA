```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# seaborn 有五种风格
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
#sns.set_style("white")
#sns.set_style("whitegrid")
#sns.set_style("dark")
sns.set_style("darkgrid") # 全局风格
#sns.set_style("ticks")
axes = sns.scatterplot(x = "total_bill",y = "tip", data = tips)
```

```python
# sns.axes_styles()
sns.axes_style()
```

```python
# 局部风格
with sns.axes_style("darkgrid", rc = {'axes.facecolor': 'y'}):# 从上面复制字典中的属性进行设置
    sns.lineplot(x = "timepoint", y = "signal", data = fmri) 
```

```python
sns.set(rc = {"lines.linewidth": 10}) # rcParams中的属性
sns.lineplot(x = "timepoint", y = "signal", data = fmri) 
```
