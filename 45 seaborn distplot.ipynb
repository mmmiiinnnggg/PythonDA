```python
# dist - distribution
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
#单变量分布
sns.set(color_codes = True)
sns.distplot(tips['tip'],kde = False,bins = 20,rug = True) #rug - 数据集中的情况 hist = False 关闭直方图
```

```python
#双变量分布
tips = sns.load_dataset("tips")
sns.jointplot(x = "total_bill",y = "tip",data = tips, kind = "reg")# reg - 回归曲线和kde曲线
```

```python
#六边形图
sns.jointplot(x = "total_bill",y = "tip",data = tips, kind = "hex",gridsize = 20,ratio = 3 )# hex - 六边形
```

```python
#成对绘图
iris = sns.load_dataset('iris')
#sns.pairplot(iris)#自动绘制数值列
sns.pairplot(tips, vars = ['total_bill','tip'])
#diag_kind 对角线图形类型
#kind : scatter reg
```
