```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
os = {
    'win7':60.86,
    'win10':18.46,
    'win8':3.61,
    'winxp':10.3,
    'mac':6.78,
    'others':1.12
}
```

```python
plt.figure(figsize=(15,10))
patches, texts, autotexts = plt.pie(os.values(),autopct = "%.2f%%", labels = os.keys(),shadow = True,explode=(0,0.1,0,0,0,0))
for autotext in autotexts:
    autotext.set_color("w")
    autotext.set_size(15)
plt.show()
```
