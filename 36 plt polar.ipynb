```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
player = ['pts','reb','ass','stl','blk',' ']#多添加一个标签
values = [90,80,75,85,95,90]#多添加一个起始点的坐标
theta = np.linspace(0,2*np.pi,6)
plt.polar(theta,values)

plt.xticks(theta,player,size=15)
plt.fill(theta,values)
plt.show()
```

```python
figure,axes = plt.subplots(1,1,subplot_kw={"projection":"polar"})
axes.plot(theta,values)
```

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_subplot(111,polar=True)
axes.plot(theta,values)
```
