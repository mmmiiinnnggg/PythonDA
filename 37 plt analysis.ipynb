```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(np.random.randn(10))

ax1.set_xlim(-2,12)#左右两边的边界
ax1.set_xticks(np.arange(0,10))#x的刻度
ax1.set_ylim(-3,3)

ax1.text(0,0,"hello")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.bar(np.arange(0,10),np.random.rand(10)*100,label = 'temp')
ax1.set_ylabel("temp")

ax2 = ax1.twinx()
ax2.plot(np.arange(0,10),np.random.rand(10),c = 'r',label = 'rain')
ax2.set_ylabel("rain")

#加图例(使他们放在一起)
fig.legend(ncol=2,loc = "upper center")#图例放几列，放在什么位置
```

```python
#axis容器
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.bar(np.arange(0,10),np.random.rand(10)*100,label = 'temp')
ax1.set_ylabel("temp")
#设置temp位置
ax1.yaxis.set_label_coords(-0.2,0.8)
#设置刻度数值
formatter = ticker.FormatStrFormatter("& %.2f")
ax1.yaxis.set_major_formatter(formatter)
#设置轴的属性
xlabels = ax1.xaxis.get_ticklabels()
for label in xlabels:
    label.set_color('r')
    label.set_rotation(30)
    label.set_fontsize(16)

#设置y轴刻度
ylines = ax1.yaxis.get_ticklines()
for line in ylines:
    line.set_color('g')
    line.set_markeredgewidth(5)
    line.set_markersize(10)
# 设置子刻度
ax1.minorticks_on()
```

```python
fig = plt.figure()
ax3 = fig.add_subplot(111)
#ticks对象：控制轴的位置 将tick放右边
for tick in ax3.yaxis.get_major_ticks():
    tick.label1On = False
    tick.label2On = True
```
