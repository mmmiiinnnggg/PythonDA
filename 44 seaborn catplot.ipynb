```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
#分类散点图 category
sns.catplot(x ="day", y = "total_bill",data = tips, hue = "sex",kind = "strip")
```

```python
sns.catplot(x ="day", y = "total_bill",data = tips,kind = "swarm")
#大数据不太适用swarm
```

```python
#分类分布图 kind = "box","violin"
athletes = pd.read_csv("active_player.csv")
teams = {
    'LAL':'Lakers',
    'GSW':'Golden State',
}
my_athletes = athletes[athletes['Team'].isin(list(teams.keys()))]
plt.figure(figsize=(5,10))
sns.boxplot(x="Team",y = "PPG",data = my_athletes)
plt.xticks(range(len(teams.values())),teams.values())
```

```python
# 小提琴图 观察数据分布 kde核密度估计曲线形成的图形
sns.violinplot(x = "Team", y = "PPG", data = my_athletes) 
# inner 中间显示的图形 默认是箱线图 inner = "quartile","point","stick"
#如果使用了hue 那么可以使用split将kde合并 更加方便比较
```

```python
# 分类统计图 barplot pointplot countplot
sns.catplot(x = "day",y = "total_bill",data = tips, kind = "bar",estimator = sum) # 线条是置信区间
#estimator - 考核方式 
```
