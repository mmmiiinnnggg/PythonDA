```python
import seaborn as sns
import pandas as pd
#load_dataset 加载默认官方数据集
tips = sns.load_dataset("tips",cache=True)#加载为df结构
tips.head()
```

```python
#relationship plot 默认是散点型函数
sns.relplot(x="total_bill",y="tip",hue = "day",col = "time",row = "sex",data = tips)#散点图
#hue 控制颜色
#col row 分割成多行多列
```

```python
sns.relplot(x="total_bill",y="tip",col="day",col_wrap=2,data = tips) #col_wrap 指的是换行
```

```python
sns.scatterplot(x="total_bill",y="tip",data = tips)#relplot默认情况下kind="scatter"
```

```python
#绘制折线图
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x = "timepoint",y = "signal",data = fmri,kind = "line",
            hue = "region",col = "event",style = "region",ci = None) #ci = None
#col - 按照指标分成两列
```

```python
import numpy as np
sns.relplot(x = "x",y= "y",data = pd.DataFrame({"x":np.arange(10),"y":np.sin(np.arange(10))}),kind = "line")
```
